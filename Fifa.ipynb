```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv("D:\python_data\greatlearning_proj\players_20.csv")
```

```python
fifa.shape
```

```python
#find the maximum number of player belongs to which nationality

fifa['nationality'].value_counts()
```

```python
# find top 5 countries with maximum no of players
player_count = list(fifa['nationality'].value_counts()[0:5])
player_count
```

```python
# find the names of the players of top 5 countries
names=list(fifa['nationality'].value_counts()[0:5].keys())
names
```

```python
# Bar plot to represent the nationality of the players

plt.figure(figsize=(5,5))
plt.bar(names,player_count,color="g")
plt.title("Nationality representation")
plt.xlabel("Countries")
plt.ylabel("player count")
plt.show()
```

```python
# Find the salary of each player
player_salary=fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending = False)
```

```python
# Arranging salary in decending order of players
player_salary.head()
```

```python
player_names_salary=list(player_salary['short_name'])[0:5]
player_names_salary
```

```python
salary = list(player_salary['wage_eur'])[0:5]
salary
```

```python
plt.figure(figsize=(7,5))
plt.bar(player_names_salary,salary,color=["blue","green","red","pink","orange"])
plt.title(" Player Salary")
plt.xlabel("player names")
plt.ylabel("salary") 
plt.show()        
        
        
```
